```python
import polars as pl
import gdown as gd
```

# **Loading Raw Data**

```python
url_events = 'https://drive.google.com/uc?id=1-aP_rI6Mhabj2X8uQ35T6bYhjcXBOSaT'
url_subs = 'https://drive.google.com/uc?id=1kFDFGq7mQYOFjVR8yFZC8E-jqWtgs44m'
url_users ='https://drive.google.com/uc?id=1-TOt6R3EHSHQXygiE7t7tOzE3GSjH764'
```

```python
gd.download(url_events,'meditrack_events_raw.csv',quiet=True)
gd.download(url_subs,'meditrack_subs_raw.csv',quiet=True)
gd.download(url_users,'meditrack_users_raw.csv',quiet=True)
```

```python
df_users = pl.read_csv('meditrack_users_raw.csv')
df_events = pl.read_csv('meditrack_events_raw.csv')
df_subs = pl.read_csv('meditrack_subs_raw.csv')
```

# **Clean Users**

```python
df_users_clean = (
    df_users
    .filter((pl.col('age').is_not_null()) & (pl.col('age') <= 80))
    .with_columns(pl.col('email').fill_null('unknown@meditrack.coom'))
)
```

# **Clean Events**

```python
valid_user_ids = df_users_clean["user_id"]
df_events_clean = df_events.filter(pl.col("user_id").is_in(valid_user_ids))
df_subs_clean = df_subs.filter(pl.col("user_id").is_in(valid_user_ids))
```

# **Save Clean Data**

```python
df_users_clean.write_csv('meditrack_users_clean.csv')
df_events_clean.write_csv('meditrack_events_clean.csv')
df_subs_clean.write_csv('meditrack_subs_clean.csv')
```
